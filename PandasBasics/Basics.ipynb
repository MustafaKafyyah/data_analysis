# Merge Data

```python
df1 = df.groupby(['Airline', 'FlightDate'])[['DepDelay']].mean().reset_index()
df2 = df.groupby(['Airline', 'FlightDate'])[['ArrDelay']].mean().reset_index()
```

```python
df1.merge(df2) # Combines similar columns so there are not dupe columns
```

```python
df1.merge(df2, how='left') # how='inner' is default
```

```python
# another way to merge
```

```python
pd.merge(df1, df2, on=['Airline']) # Can merge on specific columns
# other columns will have suffix x or y
```

```python
pd.merge(df1, df2, on=['Airline'], suffixes=('_dep','_arr'))
# can change suffix like this
```

```python
pd.merge(df1, df2, left_on=['Airline'], right_on=['Airline'], suffixes=('_dep','_arr'))
# if dataframes have different columns
```
