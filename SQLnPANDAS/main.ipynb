# Connect to DB from python 

```python
pip install pymysql
```

```python
import pymysql
```

```python
conn = pymysql.connect(
    host='localhost',
    user='root',
    passwd='',
    db='employees'
)
```

```python
curs = conn.cursor()
```

```python
query ='''
        select * 
        from employees e 
        join salaries s 
        on e.emp_no = s.emp_no 
        where e.hire_date > "1999-01-01"
        '''
```

```python
curs.execute(query)
```

```python
rows = curs.fetchall()
```

```python
results = []
for row in rows:
    results.append(row)
```

```python
conn.close()
curs.close()
```

```python
len(results)
```

```python
df_pulled = pd.DataFrame(results)
```

```python
df_pulled.columns = ['emp_no', 'birth_date', 'first_name', 'last_name', 'gender', 'hire_date', 'emp_no.1', 'salary', 'from_date', 'to_date']
```

# Writing to DB

```python
max_salary = df.groupby(['emp_no', 'first_name', 'last_name'])['salary'] \
    .max().reset_index()
```

```python
max_salary.to_sql('max_salary', engine, if_exists='replace', index=False)
```

```python
max_salary2 = pd.read_sql('SELECT * FROM max_salary', engine)
```
